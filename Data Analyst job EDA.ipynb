```python
sns.pairplot(data = data)
```

```python
sns.countplot(x='min exp', data=data)

plt.xlabel('min exp')
plt.ylabel('Count')
plt.title('Count of min exp')

plt.xticks(rotation=45)

plt.show()
```

```python
AveExp = (data['min exp'] + data['max exp']) / 2
```

```python
data.rename(columns=lambda x: x.strip(), inplace=True)

data_salaries = data.dropna(subset=['AveExp', 'AveSal'])
```

```python
sns.barplot(x='AveExp', y='AveSal', data = data_salaries)
plt.xlabel(xlabel = 'AveExp')
plt.ylabel(ylabel = 'AveSal')
plt.title(label = 'AveExp vs AveSal' )

plt.show()
```
